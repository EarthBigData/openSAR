# open_rasterio 403 error on pre-signed s3 urls

open_rasterio fails with 403 error on pre-signed s3 urls. gdal succeeds.

```python
import rioxarray as xr
from osgeo import gdal
```

```python
print(gdal.__version__)
```

```python
print(xr.__version__)
```

```python
presigned_url='https://sentinel-1-l22.s3.us-east-1.wasabisys.com/mgrs/17RNM/20m/A/048/2022/10/02/vv/17RNM_20m_A_048_2022_10_02_vv_mtfil_amp.tif?AWSAccessKeyId=LVAEZZ9KG0V0547KV4NN&Signature=8Abc5lnTO0Duex7IozgFiaYBocA%3D&Expires=1754308195'
public_url='https://ebd-public-mgrs-tiles.s3.us-east-1.wasabisys.com/mgrs/14UPU/20m/tsmetrics/COG/tile14UPU_20m_A_136_vv_mtfil_tsmetric_season_MAM_median.tif'
```

## Public urls work for both gdal and rasterio

```python
url = public_url
```

```python
print(gdal.Info(url))
```

```python
da = xr.open_rasterio(url)
da
```
